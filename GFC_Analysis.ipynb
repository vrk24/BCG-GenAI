```python
import pandas as pd

# Load your correctly named CSV file
df = pd.read_csv('Final_report.csv')

# Ensure your data is sorted correctly for percentage change calculations
# We sort by Company then Year so the math happens in the right order
df = df.sort_values(['Company', 'Year'])

# Calculate Year-over-Year growth
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100

# View the final processed data
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Pivot the data for easier plotting
pivot_df = df.pivot(index='Year', columns='Company', values='Revenue Growth (%)')

# Plotting the growth trends
pivot_df.plot(kind='line', marker='o', figsize=(10, 5))
plt.title('Company Revenue Growth Trends')
plt.ylabel('Growth Percentage (%)')
plt.grid(True)
plt.show()
```
